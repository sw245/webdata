```python
# 이름 a,A가 들어가면서 영어점수가 80점 이상인 사람 검색
filt_a = df['이름'].str.lower().str.contains('a')
filt_en = df['영어'] >=80
df[filt_a & filt_en]

## 홍길동,홍길순,유관순 검색
fsoon = df['이름'].str.endswith('순')
fdong = df['이름'].str.contains('동')
df[fsoon | fdong]

arr = ['홍길동','홍길순','유관순']
df[df['이름'].isin(arr)]
```
