```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/stuscore (1).csv',index_col='번호')
```

```python
# 합계, 평균 컬럼을 추가하고 평균 90점 미만은 데이터를 삭제
df['합계'] = df['국어'] + df['영어'] + df['수학']
df['평균'] = round( (df['합계'] / 3),2 )
df.drop(index=df[df['평균']<90].index)
```

컬럼 순서 변경

```python
# 순서 변경해서 출력후 원래 데이터에 대입하기
df = df[['이름','평균','국어','영어','수학','합계']]
```
