```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False   # 그래프 마이너스 표시
```

```python
m_df = pd.read_excel('data/2024annual.xlsx',skiprows=3,nrows=1,usecols='R:AB')
m_df = m_df.T
m_df.reset_index(inplace=True)
m_df['index'] = m_df['index'].str.slice(0,-2)
m_df['number'] = m_df[0].str.replace(',','').astype(int)
```

```python
w_df = pd.read_excel('data/2024annual.xlsx',skiprows=3,nrows=1,usecols='AE:AO')
w_df = w_df.T
w_df.reset_index(inplace=True)
w_df['index'] = w_df['index'].str.slice(0,-2)
w_df['number'] = w_df[0].str.replace(',','').astype(int)
```

```python
pops = np.arange(-4500000,4500000,1000000)
```

```python
m_df
w_df
```

```python
plt.figure(figsize=(20,7),dpi=300)
plt.title('연령 별 인구 현황')
plt.barh(w_df['index'],m_df['number'])
plt.barh(w_df['index'],w_df['number']*(-1))
plt.xticks(pops,rotation=45)
plt.show()
```
