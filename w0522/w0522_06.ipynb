```python
# plt.bar(df['이름'],df['국어'])   


# label >> 범례(그래프 구분)
# marker : 지점에 마커표시 - o,v,x,s 등
# markersize - ms, markeredgecolor - mec, markerfacecolor - mfc
# linestyle - ls: 선 모양 지정 - ( None , : , -- , -. ) 등
# linewidth: 선 두께, color: 선 색 지정
# alpha: 선 투명도 조절

# plt.plot(x,y,label='서울',linewidth=1.5,ls = '-.',alpha=0.7,color='plum',
#          marker='*',ms=11,markeredgecolor='indigo',markerfacecolor='pink')   # 꺾은선 그래프


# 축약형: 'color,마커,선' >> 'ro--' ('r''o''--')
plt.plot(x,y,'bv--',label='서울',linewidth=1.3)

plt.title('그래프')     # 한글지원 설정해야함
plt.legend(loc='upper left')   
# loc = upper,center,lower / left, center, right >> 범례 위치 지정
# plt.legend(loc=(0.25,0.5))

# plt.show()  # 실행구문 제거
```

```python
plt.plot(df['이름'],df['국어'],label='국어',marker='o',linestyle='None')
plt.plot(df['이름'],df['수학'],label='수학',marker='v',linestyle='None')
plt.plot(df['이름'],df['영어'],label='영어',marker='X',linestyle='None')
plt.title('성적그래프')
plt.legend()
plt.show()
```
