```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False   # 그래프 마이너스 표시
```

```python
day = [1,2,3]
kor = [100,80,90]
eng = [95,85,92]
math = [80,100,70]
```

그래프 여러 개 그리기

```python
plt.plot(day,kor,label='국어',marker='o')
plt.plot(day,eng,label='영어',marker='v',ls='--')
plt.plot(day,math,label='수학',marker='X',ls=':')
plt.xticks([1,2,3])
plt.yticks([60,70,80,90,100])
plt.legend(ncol=3)  # ncol: 숫자만큼 한 행으로 합쳐서 보여줌
for i,v in enumerate(kor):
    plt.text(day[i],kor[i]-2.5,v,ha='center')
    plt.text(day[i],eng[i]+1,eng[i],ha='center')
    plt.text(day[i],math[i]-2.5,math[i],ha='center')
plt.show()
```

```python
name = ['홍','유','이']
kor = [100,80,90]
eng = [95,85,92]
math = [80,100,70]
```

```python
# no = day
no = np.array([1,2,3])
```

```python
w = 0.3
bar = plt.bar(no-w,kor,color='red',label='국어',alpha=0.6,width=0.3)
## .set_hatch : bar에 무늬 적용
bar[0].set_hatch('/')   
bar[1].set_hatch('x')   
bar[2].set_hatch('..')  

plt.bar(no,eng,color='green',label='영어',alpha=0.6,width=0.3)
plt.bar(no+w,math,color='blue',label='수학',alpha=0.6,width=0.3)
plt.ylim(40,115)
plt.yticks([60,70,80,90,100])
for i,v in enumerate(kor):
    plt.text(no[i]-w,kor[i]+0.5,v,ha='center')
    plt.text(no[i],eng[i]+0.5,eng[i],ha='center')
    plt.text(no[i]+w,math[i]+0.5,math[i],ha='center')
plt.legend(ncol=3)
# width: 바 너비 조정
# no >> numpy 배열 형태?
plt.xticks(no,name,rotation=20)     # (쓰는 값,표기 값) - x좌표 이동에 필요한 수치를 no로 사용 가능
```
