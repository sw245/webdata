```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.unicode_minus'] = False   # 그래프 마이너스 표시
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,'ro--',ms=8,mec='red',mfc='yellow',linewidth=7,label='y')
plt.title('그래프')
plt.xlabel('월별')
plt.xticks([1,2,3])

plt.legend()
# plt.show()
```

```python
df = pd.read_csv('data/stuscore (1).csv')
df3 = df.head(3)

# 영어 꺾은선 그래프 점수 표시
```

```python
plt.plot(df3['이름'],df3['영어'],label=df3.columns[3])
plt.title('영어 성적')
plt.legend()
plt.yticks([50,70,90,100])

for i,s in enumerate(df3['영어']):
    plt.text(df3['이름'][i],df3['영어'][i]+1,s,ha='center')
```
