데이터 선택 loc > row에 대한 선택

```python
import pandas as pd
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
```

```python
## 컬럼 출력
df['이름']
df[['이름','학교']]
df.columns
df[df.columns[:3]][1:3]
df[[df.columns[-1]]]
```

```python
## row 출력 - loc: 인덱스 이름을 가지고 출력
df.loc[['1번']]
df[0:2]
df.loc[['1번','2번']]   # == df[0:2]

## row,column 동시 지정 출력
df.loc['1번']['이름']  
df['이름'].loc['1번']

df.loc[['1번','2번'],['국어','영어']]
df.loc[['1번','2번']][['영어','수학']]
df.loc['1번',['국어','영어']]

## row, column 슬라이싱 출력
df.loc['1번':'5번','국어':'사회']
```

```python
## row 출력 - iloc: 인덱스 번호로 출력

df.iloc[0]
df.iloc[[4]]
df.iloc[0:4]

# 행의 기준으로 열을 출력 
df.iloc[0,1]    # == df.loc['1번','학교']
df.iloc[[0,1,3],1]  # == df.loc[['1번','2번','4번'],'학교']

df.iloc[[0,1],[3,4]]    # 0,1행, 3,4열 출력


# 슬라이싱
df.iloc[:4,3:]
df.iloc[1:4,3:7]    # 1,2,3행, 3,4,5,6열 출력

```
