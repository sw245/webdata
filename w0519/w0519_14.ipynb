```python
# str 함수
df3 = pd.read_excel('data/people.xlsx',skiprows=2,usecols='B:G',index_col='행정기관')
```

```python
# str.replace() 함수 - 문자 바꿔넣기

df3.columns
df3['총인구수'] = df3['총인구수'].str.replace(",",'').astype(int)
```

```python
## str.startswith() 함수 - 해당 문자로 시작되는 항목 검색
filt = df2['이름'].str.startswith('이')
filt = df2['이름'].str.startswith('S')
df2[filt]
```

```python
## str.endswith() 함수 - 해당 문자로 끝나는 항목 검색
df2[df2['이름'].str.endswith('순')]
df2[df2['이름'].str.endswith('n')]
```

```python
## str.contains() - 해당 문자가 포함된 항목 검색
df2[df2['이름'].str.contains('e')]
df2[df2['이름'].str.contains('신')]
df2[~df2['이름'].str.contains('순')]    # ~ >> not
```

```python
## a가 들어가 있는 이름 항목을 검색
df2[df2['이름'].str.contains('a')]
df2[df2['이름'].str.contains('a')].count()
df2[df2['이름'].str.contains('A')].count()
df2[df2['이름'].str.contains('A')]
df2[df2['이름'].str.contains('A')|df2['이름'].str.contains('a')]
df2[df2['이름'].str.contains('A')|df2['이름'].str.contains('a')].count()


# str.lower() - 모든 글자 소문자로 변환
# str.upper() - 모든 글자 대문자로 변환
df2[df2['이름'].str.lower().str.contains('a')]
df2[df2['이름'].str.lower().str.contains('a')].count()
df2[df2['이름'].str.upper().str.contains('A')]
```
