```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
```

```python
# int타입에 cm문자 추가
# df['키'] = df['키'].astype(str) + 'cm'

# 함수 사용해서 추가
def add_cm(height):
    return str(height) + 'cm'

# add_cm(df['키'])    # >> 이거 아님...

df['키'] = df['키'].apply(add_cm)
```

```python
df['합계'] = df['과학'] + df['국어'] + df['사회'] + df['수학'] + df['영어']
df['등급'] = 'F'
df['평균'] = df['합계'] / 5


# 함수를 사용해서 90점 이상 합격, 90점 미만 불합격
df['결과'] = ''
def passornot(avg):
    if avg>=90:
        return '합격'
    else:
        return '불합격'
    
df['결과'] = df['평균'].apply(passornot)



# 90점 이상 A, 80점 이상 B, C,D,F

def grade(avg):
    if avg>=90:
        return 'A'
    elif avg>=80:
        return 'B'
    elif avg>=70:
        return 'C'
    elif avg>=60:
        return 'D'
    else:
        return 'F'
    
df['등급'] = df['평균'].apply(grade)
```

```python
# SW특기의 글자를 모두 소문자로 변경

df['SW특기'] = df['SW특기'].str.lower()
df['SW특기'].str.capitalize()
```

```python
# java라는 문자가 포함된 항목 출력

df.loc[df['SW특기'].str.contains('java',na=False)]
```

```python
# 함수를 사용해서 SW특기를 전부 대문자로 변경하고, NaN 데이터는 C를 입력


## isna() >> pandas함수 // isnull(),notnull() >> pandas 함수 ??
def fillSW(x):
    if pd.isna(x):
        return 'C'
    else:
        return x.upper()
        
df['SW특기'] = df['SW특기'].apply(fillSW)
```

```python
## 문자열 슬라이싱
df['학교'].str.slice(0,-1)

# df['키'] = df['키'].astype(str) + 'cm'
# df['키'] = df['키'].str.replace('cmcm','cm')
# df['학교'] = df['학교'] + '등학교'

df['학교'].str.slice(0,-3)
df['키'].str.slice(0,-2)
```

```python
# zfill: 자릿수만큼 빈 공간을 0으로 채움 (문자열만 가능, 숫자타입 X)
df['이름'].str.zfill(width=4)
df['국어'].astype(str).str.zfill(width=3)   
```

```python
# df['요일'] = 0

for i in range(8):
    df.iloc[i,13] = random.randint(0,6)


# 함수를 사용해서 요일에 0:일요일, 1:월요일 ... 할당

def days(x):
    if x==0: return '일요일'
    elif x==1: return '월요일'
    elif x==2: return '화요일'
    elif x==3: return '수요일'
    elif x==4: return '목요일'
    elif x==5: return '금요일'
    elif x==6: return '토요일'
    
df['요일'] = df['요일'].apply(days)
df
```
