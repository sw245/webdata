```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
```

```python
## 그룹화
df.groupby('학교')
df.groupby('학교').get_group('가산고')
df.groupby('학교').get_group('디지털고')
# 숫자형만 통계함수 적용 가능
df.groupby('학교').size()       # 항목이 몇 개 있는지
df.groupby('학교').size()['가산고']     # '가산고' 그룹이 몇 개 항목인지
df.groupby('학교')['키'].mean()     # 학교 별 평균 키
df.groupby('학교')[['키','국어']].mean()     # 학교 별 평균 키, 평균 국어점수
df.groupby('학교')[['키','국어','영어']].max()  # 학교 별 각 컬럼에서의 최댓값
df.groupby('학교')[['키','국어','영어']].min() 
df.groupby('학교')[['키','국어','영어']].sum() 

# 숫자가 아닌 타입이 있어도 가능
df.groupby('학교').count()
df['학교'].count()
df['학교'].value_counts()           # 학교 그룹 별 개수
df.groupby('학교')['학교'].count()   # 동일함 (학교 그룹 별 개수)

df.groupby('학교')['학년'].count().loc['가산고']    # 가산고 항목 개수

df.groupby('학교').value_counts()
df.groupby('학교')['학년'].value_counts()   # 1차 학교로 그룹, 2차 학년으로 그룹
df.groupby(['학교','학년'])['학년'].count()

```

```python
# df.groupby('학교')[['이름','SW특기']].count()
# df.groupby('학교').value_counts()
# df['학교'].value_counts()
# df.groupby('학교')['학년'].value_counts()

# agg() 기본적인 통계정보를 알려줌 - 평균,최소,최대,합계
df.groupby('학교')['키'].agg(['mean','min','max','sum'])    # 학교 그룹의 키 컬럼 통계
df['키'].agg(['mean'])      # 키 컬럼 전체 평균
df.groupby('학년')['키'].agg(['mean'])  
```

```python
# df['성별'] = ['남','여','남','남','남','여','여','여']
df['성별'] = ['남','여','여','남','남','여','여','남']
# df['성별'] = ['남','여','남','여','남','여','남','여']
```

```python
# 합계 컬럼 추가
# 학교, 성별 그룹으로 합계컬럼의 평균,최소,최대,합계점수 

# df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']

df.groupby(['학교','성별'])['합계'].agg(['mean','min','max','sum','count'])
```

```python
# df에 '학년' 추가
df['학년'] = [3,3,2,1,1,3,2,2]

# 학년 별 평균 - 키,국어,영어 출력
df.groupby('학년')[['키','국어','영어']].mean().round(2)


# 1차 그룹 - 학교, 2차 그룹- 학년
df.groupby(['학교','학년'])[['키','국어','영어']].mean().round(2)

df.groupby('학년')['키'].mean().sort_values()
df.groupby('학년')[['키','영어','국어']].mean().sort_values(['키','국어'],ascending=False)
```
