```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
```

```python
# '홍' 포함된 이름 삭제 후 출력
df.drop(index=df[df['이름'].str.contains('홍')].index)


# 합계를 구하고 400점 미만 삭제
# df['합계'] = df['과학'] + df['국어'] + df['사회'] + df['수학'] + df['영어']
df.drop(index=df[df['합계']<400].index)

# 합계, 결과 컬럼 추가
# df['결과'] = '불합격'

# 400점 이상 합격, 400점 미만 불합격
# df.loc[df['합계']>=400,'결과'] = '합격'

# 컬럼 순서 변경 >> 이름,결과,합계,키 ... 순
col = list(df.columns)
df.columns = [col[0]] + [col[-1]] + [col[-2]] + col[1:-2]
```
